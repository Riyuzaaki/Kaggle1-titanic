```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
import math
x_train=pd.read_csv('train.csv')
x_test=pd.read_csv('test.csv')
```

```python
x_train.drop('Cabin',axis=1,inplace=True)
```

```python
sex=pd.get_dummies(x_train['Sex'],drop_first=True)
sex
```

```python
sex2=pd.get_dummies(x_test['Sex'],drop_first=True)
sex2
```

```python
embark=pd.get_dummies(x_train['Embarked'],drop_first=True)
embark
```

```python
embark2=pd.get_dummies(x_test['Embarked'],drop_first=True)
embark2
```

```python
pcl= pd.get_dummies(x_train['Pclass'],drop_first=True)
pcl
```

```python
pcl2= pd.get_dummies(x_test['Pclass'],drop_first=True)
pcl2
```

```python
x_train=pd.concat([x_train,sex,embark,pcl],axis=1)
```

```python
x_test=pd.concat([x_test,sex2,embark2,pcl2],axis=1)
```
